```python
# Q1) Given a list of 2-D coordinates, write a function that returns True
#     if the points lie on a straight line and False otherwise.
#
#     Use only the Python standard libraries.
#
# collinear([[1,1], [2,2], [4,4], [-10, -10]]) = True
# collinear([[1,0], [2,0], [3,1]]) = False

def collinear(points):
    #If there are only 2 points it will always be a line
    if len(points) < 2:
        return True
    #Use the first two points to find the slope and intercept
    x0, y0 = points[0]
    x1, y1 = points[1]
    slope = (y1 - y0) / (x1 - x0)
    intercept = y0 - (x0*slope)
    #Check if subsequent points are on the same line
    for i in range(2, len(points)):
        x,y = points[i]
        if y - (slope * x) == intercept:
            return True
        else:
            return False

# Test cases
print(collinear([[1,1], [2,2], [4,4], [-10, -10]])) # True
print(collinear([[1,0], [2,0], [3,1]])) # False
print(collinear([[1,1], [2,2], [4,4], [-10, -10], [17,17]]))
```

```python
# Q2) Define a function that returns the frequencies of the last digits
#     of a list of nonnegative integers.
#
#     Use only the Python standard libraries.
#
#     Given the list [49, 10, 20, 5, 30, 785]:
#     9 is the last digit once (in 49),
#     0 is the last digit three times (in 10, 20 and 30),
#     5 is the last digit two times (in 5 and 785)
# last_digit_counts([49, 10, 20, 5, 30, 785])
#   = {9:1, 0:3, 5:2} # or something equivalent

def last_digit_counts(numbers):
    counts = {}
    #Convert to string, extract the last number then increment the last digit count if it already exists in the dict
    for number in numbers:
        last_digit = int(str(number)[-1])
        if last_digit in counts:
            counts[last_digit] += 1
        else:
            counts[last_digit] = 1
    return counts

#test cases
numbers = [49, 10, 20, 5, 30, 785]
print(last_digit_counts(numbers))
numbers = [35, 24, 14, 47, 84, 56, 98, 1, 104, 2943, 54]
print(last_digit_counts(numbers))
```

```python
# Q3) Using whatever library you like, make an effective
#     visualization for the data stored here:
#     https://github.com/michaelbilow/open-data/raw/main/spotify-2023.xlsx
#     It consists the most streamed songs on Spotify in 2023,
#     along with their artist, key, mode, and year of publication.
#     Write 1-2 sentences about what you're trying to show.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'spotify-2023.xlsx'
df = pd.read_excel(file_path)

#Chart1: Find BPM buckets with the most streams
# Create BPM buckets
df['BPM Bucket'] = (df['BPM'] // 10) * 10

# Group by BPM Bucket and sum the Streams
bpm_streams = df.groupby('BPM Bucket')['Streams'].sum().reset_index()

# Create a new column for bucket labels
bpm_streams['Bucket Label'] = bpm_streams['BPM Bucket'].astype(str) + ' - ' + (bpm_streams['BPM Bucket'] + 9).astype(str)

# Set Seaborn style
sns.set(style="whitegrid")

# Create the bar chart with Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Bucket Label', y='Streams', hue='Bucket Label', data=bpm_streams, palette='viridis', dodge=False, legend=False)
plt.xlabel('BPM Bucket')
plt.ylabel('Total Streams')
plt.title('Total Streams by BPM Bucket')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()


#Chart 2: Find Keys with the most streams
# Group by Key and sum the Streams
key_streams = df.groupby('Key')['Streams'].sum().reset_index()

# Sort by Streams in descending order
key_streams = key_streams.sort_values(by='Streams', ascending=False)

# Set Seaborn style
sns.set(style="whitegrid")

# Create the bar chart with Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Key', y='Streams', hue='Key', data=key_streams, palette='viridis', legend=False)
plt.xlabel('Key')
plt.ylabel('Total Streams')
plt.title('Total Streams by Key')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()


#Chart 3: Pie chart showing Mode distribution
# Pie Chart of Mode Distribution
mode_counts = df['Mode'].value_counts()
plt.figure(figsize=(8, 8))
mode_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Mode Distribution')
plt.ylabel('')
plt.show()


#Chart 4: Find the top 10 Artists with the most unique songs in the file
# Count the number of songs each artist has in the list
artist_song_counts = df['Artist'].value_counts().reset_index()
artist_song_counts.columns = ['Artist', 'Song Count']

# Sort the artists by the number of songs in descending order and select the top ten
top_ten_artists = artist_song_counts.sort_values(by='Song Count', ascending=False).head(10)

# Set Seaborn style
sns.set(style="whitegrid")

# Create the bar chart with Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Artist', y='Song Count', data=top_ten_artists, palette='viridis', hue='Artist', legend=False)
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.title('Top 10 Artists by Number of Songs in the List')
plt.xticks(rotation=90)
plt.grid(axis='y')

plt.show()
```

Chart 1: I wanted to know what BPMs make for the most popular songs. I bucketed them into groups of 10 BPMs then summed the total streams in the bucket.

Chart 2: Bar chart showing which key has the most streams

Chart 3: Pie chart breakout between major and minor mode

Chart 4: Bar chart to find which artists have the most unique songs in the list (solos only, does not includes songs with feature)

### Q4) If you flip a fair coin 100 times (independently), what is the chance that more than 60 flips come up heads?

**Explanation**:
- The standard deviation given 100 flips (50% chance for heads each flip) is:
$$
\sigma = \sqrt{100 \cdot 0.5 \cdot 0.5} = 5
$$
- So more than 60 is 2 standard deviations away from mean 50.
- 2 standard deviations is ~5% of the normal distribution.
- 5% split between right and left tails, so divide by 2 to get ~2.5%.

**Conclusion**:
The probability of getting more than 60 heads out of 100 flips of a fair coin is approximately 2.5%.
